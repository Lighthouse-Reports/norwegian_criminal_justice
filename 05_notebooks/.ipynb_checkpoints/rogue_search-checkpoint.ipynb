### **Aggregate data**

**Global filter.** We're only interested in bias among convictions, so we exclude non-convictions.

**At a glance.** We can see how the courts and judges are distibuted. For courts, the uneven distribution suggests that we might want to set a threshold for the minimum number of occurrences before doing any final analysis. For judges, **TK**

**Benchmarks.** Pulling representatitve statistics from the aggregate dataset will give us benchmarks to test our disaggregated datasets against for bias. We'll use:
> 1. The first, second (i.e., the median), and third quartiles of each punishment.
> 2. The interquartile range of each punishment as well.
> 3. **TK**

We'll pull these stats from three variables measured in [Data Memo 2.0 (08/24)](https://docs.google.com/document/d/1xZ_N1XgKHe8Wof5ZyVFogcq7sGrnN16w2iyl3hLRDUc/edit?usp=sharing):

> 2. `punishment_prison_days`
> 3. `punishment_share_max_prison_sentence`

**First aggregate filter.** We'll remove NaN and 0 values before pulling any stats out. Not only do these values lower any statistical measures we calculate here, they also correspond to cases where the judges did *not* hand out punishments. It arguably makes more sense to look for outlier courts only among those cases where punishments were handed out, and thus where their bias could manifest.

```python
def quart(df : pd.DataFrame, colName : str) -> np.array:
    ''' Finds quartiles and interquartile range of given column from a DataFrame '''
    df.dropna(subset=[colName],inplace=True) #exlcude NaN values
    df=df[df[colName]!=0] #exclude 0 values
    return np.nanpercentile(df[str(colName)].values, [25, 50, 75])

punishment_fine_quart = quart(df, "punishment_fine")
punishment_prison_days_quart = quart(df, "punishment_prison_days")
punishment_share_max_prison_sentence_quart = quart(df, "punishment_share_max_prison_sentence")
```

### **District courts**

First, we create a list of the courts.

**Matching on judge ID.** By filtering `df` for each unique court code, we can selectively look at each district court's cases.

**First disaggregate filter.** Like with aggregate data, we remove NaN and 0 values.

**Second disaggregate filter.** Filters for convictions and demographics **TK**.

Now we check the median of each punishment against the IQR of the benchmark.

A *high outlier* is defined as: $f(\text{court}) > Q3 + 1.5*IQR$.

A *low outlier* is defined as:  $f(\text{court}) < Q1 - 1.5*IQR$.

**Third disaggregate filter.** We can see large discrepancies between the number of outliers for each punishment. In order to find judges that are biased irrespective of which case they try, we'll match on all punishments.

```python
#Drop NaN values in the court codes
df.dropna(subset='court_code', inplace=True)
#Generate list of unique court codes
codes = df['court_code'].unique()
```

```python
# We have to define a special function to ignore NaN and 0 values 
def median(series : pd.Series) -> float:
    series = series.dropna() #Drop NaN values
    series = series[series != 0] #Drop 0 values
    if series.empty: #Manual fix to ensure empty series don't return NaN as the median
        return 0
    else:
        return series.median()

courts = df.groupby('court_code').agg(
    #Fill new columns
    number_of_cases=('court_code', 'size'),
    median_punishment_fine=('punishment_fine', lambda x: median(x)),
    median_punishment_prison_days=('punishment_prison_days', lambda x: median(x)),
    median_punishment_share_max_prison_sentence=('punishment_share_max_prison_sentence', lambda x: median(x))
).reset_index()
```

```python
def biascheck(colname : pd.Series, series : pd.Series) -> bool:
    quart = globals()[f'{colname}_quart'] #call quartile array
    iqr = quart[2] - quart[0]
    if series > quart[2] + iqr:
        return 'Higher'
    if series < iqr - quart[0]:
        return 'Lower'
    else:
        return None
        
courts['punishment_fine_bias'] = courts.apply(lambda row: biascheck('punishment_fine', row['median_punishment_fine']), axis=1)
courts['punishment_prison_days'] = courts.apply(
    lambda row: biascheck('punishment_prison_days', row['median_punishment_prison_days']), axis=1)
courts['punishment_share_max_prison_sentence'] = courts.apply(
    lambda row: biascheck('punishment_share_max_prison_sentence', row['median_punishment_share_max_prison_sentence']), axis=1)
```

```python
#courts
```

```python
cond1_lower = courts['punishment_fine_bias']=='Lower'
cond2_lower = courts['punishment_prison_days']=='Lower'
cond3_lower = courts['punishment_share_max_prison_sentence']=='Lower'

courts_lower = courts[cond1_lower&cond2_lower&cond3_lower]
print(f"There are {len(courts_lower)} courts which are outliers below the median of all three punishments")

cond1_higher = courts['punishment_fine_bias']=='Higher'
cond2_higher = courts['punishment_prison_days']=='Higher'
cond3_higher = courts['punishment_share_max_prison_sentence']=='Higher'

courts_higher = courts[cond1_higher&cond2_higher&cond3_higher]
print(f"There are {len(courts_higher)} courts which are outliers above the median of all three punishments")
```

#### **Hypothesis 2: There's a correlation between punishments and the number of times a lay judge has overseen a case.** 

Thinking: repeat lay judges will be more familiar with the process and this may affect the punishments they hand out.

```python
#Add counts of cases to larger df
counts = lay_judges['lay__id'].value_counts()
lay_judges["number_of_cases_for_id"] = lay_judges['lay__id'].map(counts)
```

```python
#Finding stats
def stat_case(case, column):
    cases = lay_judges[lay_judges['number_of_cases_for_id'] == case]
    return median(cases[column])

#Histogram suggests only IDs with 1-3 cases worth keeping
fine = (stat_case(1, 'punishment_fine'), stat_case(2, 'punishment_fine'), stat_case(3, 'punishment_fine'))
prison = (stat_case(1, 'punishment_prison_days'), stat_case(2, 'punishment_prison_days'), 
          stat_case(3, 'punishment_prison_days'))
share = (stat_case(1, 'punishment_share_max_prison_sentence'), stat_case(2, 'punishment_share_max_prison_sentence'),
         stat_case(3, 'punishment_share_max_prison_sentence'))
```

```python
#Plotting
plt.scatter(lay_judges['number_of_cases_for_id'], lay_judges['punishment_fine'], marker='.')
plt.scatter((1, 2, 3), (fine[0], fine[1], fine[2]), color='red', label='Median')
plt.xlim(0, 3.5)
plt.legend()
plt.show()

plt.scatter(lay_judges['number_of_cases_for_id'], lay_judges['punishment_prison_days'], marker='.')
plt.scatter((1, 2, 3), (prison[0], prison[1], prison[2]), color='red', label='Median')
plt.xlim(0, 3.5)
plt.show()

plt.scatter(lay_judges['number_of_cases_for_id'], lay_judges['punishment_share_max_prison_sentence'], marker='.')
plt.scatter((1, 2, 3), (share[0], share[1], share[2]), color='red', label='Median')
plt.xlim(0, 3.5)
plt.show()
```

No correlation seems present.
